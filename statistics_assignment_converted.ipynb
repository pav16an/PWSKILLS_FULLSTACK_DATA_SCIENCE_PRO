# Statistics Practice Assignment
This notebook contains implementations of various statistical functions and visualizations.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

## Generate a random list of integers
We generate a list of 100 integers between 90 and 130.

```python
int_list = np.random.randint(90, 130, size=100)
int_list
```

## 1. Mean and Median
Functions to calculate mean and median of a list.

```python
mean_value = np.mean(int_list)
median_value = np.median(int_list)
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
```

## 2. Mode
Calculate the mode of the list.

```python
from scipy import stats

int_list = [1, 2, 2, 3, 4]
mode_value = stats.mode(int_list)
print(f"Mode: {mode_value.mode}, Count: {mode_value.count}")
```

## 3. Weighted Mean
Calculate weighted mean assuming equal weights.

```python
def weighted_mean(values, weights):
    total_weight = sum(weights)
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    return weighted_sum / total_weight

weights = np.ones_like(int_list)  # equal weights
wm = weighted_mean(int_list, weights)
print(f"Weighted Mean (equal weights): {wm}")
```

## 4. Geometric Mean
Calculate geometric mean of the list.

```python
from scipy.stats import gmean
geo_mean = gmean(int_list)
print(f"Geometric Mean: {geo_mean}")
```

## 5. Harmonic Mean
Calculate harmonic mean of the list.

```python
def harmonic_mean(numbers):
    n = len(numbers)
    reciprocal_sum = sum(1 / num for num in numbers)
    return n / reciprocal_sum

harm_mean = harmonic_mean(int_list)
print(f"Harmonic Mean: {harm_mean}")
```

## 6. Midrange
Calculate midrange (average of min and max).

```python
def mid_range(numbers):
    return (min(numbers) + max(numbers)) / 2

midrange_value = mid_range(int_list)
print(f"Midrange: {midrange_value}")
```

## 7. Trimmed Mean and Outliers
Identify outliers using IQR and Z-score methods.

```python
import numpy as np

int_list = [1, 2, 3, 4, 100]  # Example list with an outlier
int_array = np.array(int_list)

Q1 = np.percentile(int_array, 25)
Q3 = np.percentile(int_array, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = int_array[(int_array < lower_bound) | (int_array > upper_bound)]
print(f"Outliers using IQR method: {outliers_iqr}")
```

## 8. Generate new list and visualize
Generate a list of 500 integers between 200 and 300 and visualize frequency, Gaussian distribution, and KDE plots.

```python
int_list2 = np.random.randint(200, 300, size=500)

# Frequency distribution
plt.figure(figsize=(10,6))
sns.histplot(int_list2, bins=30, kde=False)
plt.title('Frequency Distribution')
plt.show()

# Gaussian distribution
plt.figure(figsize=(10,6))
sns.histplot(int_list2, bins=30, kde=True)
plt.title('Gaussian Distribution with KDE')
plt.show()

# KDE plot
plt.figure(figsize=(10,6))
sns.kdeplot(int_list2, shade=True)
plt.title('Kernel Density Estimate (KDE) Plot')
plt.show()
```
