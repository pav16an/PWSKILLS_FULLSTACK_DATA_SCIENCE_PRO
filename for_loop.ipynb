#10. Develop a program that prints the first `n` Fibonacci numbers using a `for` loop.

```python
def fibonacci(n):
    fib_sequence = [0,1]

    for _ in range(2,n):
        next_number = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_number)

    return fib_sequence


n= int(input("Enter the value of n: "))

fibonacci_numbers = fibonacci(n)
print(f"The first {n} Fibonacci numbers  are :",fibonacci_numbers)
```

11. Write a program to check if a given list contains any duplicates using a `for` loop.

```python
def duplicates(input_list):
    seen = set()

    for item in input_list:
        if item in seen:
            return True
        seen.add(item)

    return False
my_list = [1,2,3,4,5,6,4,322,4]
if duplicates(my_list):
    print("The list contains duplicates.")
else:
    print(" The list does not contain duplicates")
```

14. Write a program to find the maximum element in a 2D list using a nested `for` loop.

```python
def max_element(matrix):
    if not matrix or not matrix[0]:
        return None
    
    max_element = matrix[0][0]

    for row in matrix:
        for element in row:
            if element > max_element:
                max_element = element
    return max_element

matrix = [
    [1,2,3],
    [9,5,6],
    [7,8,79]
]

Max_element = max_element(matrix)

if max_element is not None:
    print("The maximum element in the 2D list is:",Max_element)
else:
    print("The 2D list is empty")

```

7. Write a program that converts a list of Fahrenheit temperatures to Celsius using a `for` loop.

```python
def fahrenheit_to_celsius(fahrenheit_temps):
    celsius_temps = []

    for fahrenheit in fahrenheit_temps:

        celsius = (fahrenheit - 32) * 5/9

        celsius_temps.append(celsius)

    return celsius_temps


fahrenheit_temps = [32,68,104,212]
celsius_temps_result = fahrenheit_to_celsius(fahrenheit_temps)

print(f"Fahrenheit temperatures: {fahrenheit_temps}")
print(f"Celsius Temperatures: {celsius_temps_result}")
```

31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list comprehension.

```python
def sieve_of_eratosthenes(limit):
    primes = [True] * (limit +1)
    primes[0] = primes[1] = False 

    for i in range(2, int(limit**0.5) + 1):
        if primes[i]:
            primes[i*i: limit+1:i] = [False] * len(primes[i*i:limit+1:i])

    return [ num for num in range(2, limit + 1) if primes[num]]

limit = 50
prime_numbers = sieve_of_eratosthenes(limit)
print("Prime numbers up to", limit, "are:",prime_numbers)
```

32. Create a program that generates a list of all Pythagorean triplets up to a specified limit using list comprehension

```python
def pythagorean_triplets(limit):
    triples = [(a,b,c) for a in range(1, limit+1)
               for b in range(a, limit+1)
               for c in range(b, limit+1)
               if a**2 + b**2 == c**2]
    return triples

limit = 20
pythagorean_triplets_list = pythagorean_triplets(limit)
print("Pythagorean triplets up to", limit , "are :", pythagorean_triplets_list)
```

34. Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.

```python
from statistics import mean,median,mode

def calculate_statistics(numbers):
    mean_value= mean(numbers)
    median_value = median(numbers)
    try:
        mode_values = mode(numbers)
    except StatisticsError:
        mode_values = "No unique mode"

    return mean_value, median_value, mode_values



number_list = [1,2,3,4,5,6,7,8,9]
mean_result, median_result, mode_result = calculate_statistics(number_list)

print('Mean:', mean_result)
print('Median:',median_result)
print('Mode:',mode_result)
```

35. Create a program that generates Pascal's triangle up to a specified number of rows using list comprehension.

```python
def generate_pascals_triangle(num_rows):
    triangle = [[1] * (i + 1) for i in range(num_rows)]
    
    for i in range(2, num_rows):
        for j in range(1, i):
            triangle[i][j] = triangle[i-1][j-1] + triangle[i-1][j]
    
    return triangle

# Example: Generate Pascal's Triangle with 5 rows
num_rows = 5
pascals_triangle = generate_pascals_triangle(num_rows)

# Print the generated Pascal's Triangle
for row in pascals_triangle:
    print(row)
```

36. Develop a program that calculates the sum of the digits of a factorial of numbers from 1 to 5 using list comprehension.

```python
from math import factorial

def sum_of_digits_factorials(limit):
    return [sum(int(digit) for digit in str(factorial(i))) for i in range(1, limit + 1)]

# Example: Calculate the sum of digits of factorials for numbers 1 to 5
limit = 5
sum_of_digits_list = sum_of_digits_factorials(limit)

# Print the results
for i, result in enumerate(sum_of_digits_list, start=1):
    print(f"The sum of digits for {i}! is {result}")
```

37. Write a program that finds the longest word in a sentence using list comprehension

40. Write a program that generates a list of prime palindromic numbers using list comprehension.
