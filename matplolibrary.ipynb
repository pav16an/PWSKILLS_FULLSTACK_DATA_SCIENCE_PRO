```python
x=np.linspace(1,10,100)# to print the numbers 100 inbetween 1,10
x
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x), "-_" )
plt.show()
#plt.figure() creates a new figure.
#plt.plot() adds a line plot to the figure.
#plt.show() displays the figure with the plot.
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.figure()
plt.subplot(1,2,1)# is is makes the figure into two parts and shows the first part
plt.plot(x,np.sin(x))

plt.subplot(1,2,2)# it is shows the second part of the figure
plt.plot(x,np.cos(x))

plt.show()

```

```python
#>>subplot
# function takes three integers as parameters:
#nrows: The number of rows in the grid.
#ncols: The number of columns in the grid.
#index: The position of the subplot in the grid, numbered from 1 to nrows * ncols
```

```python
plt.figure()
plt.subplot(2,2,1)# it shows the first part of two parts of the figure
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x, np.cos(x))

plt.subplot(2,2,4)
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='green')
plt.plot(x, np.sin(x-1), color='blue')
plt.plot(x, np.sin(x-2), color='green')
plt.plot(x, np.sin(x-3), color='pink')
plt.plot(x, np.sin(x+1), color='blue')
plt.show()
```

```python
plt.plot(x, np.sin(x-1), color='blue')
plt.ylim(-1,1.5)
plt.xlim(1,15)
plt.grid()# it is adds the grids to the graphe
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x-1), color='blue', label='Sinx')
plt.plot(x, np.cos(x-1), color='red', label = 'Cosx')
plt.ylim(-1,1.5)
plt.xlim(1,15)
plt.grid()# it is adds the grids to the graphe
plt.title('This is sin plot before real dataset execution')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()# it is makes the labels visible inside the graphe
plt.show()
```

```python
x = np.random.randn(50)
y =np.random.rand(50)
plt.figure(figsize = (6,3))
plt.scatter(x,y,c= 'r' , alpha= 0.8,marker='*')# alpha is used to manage the intensity of the color and marker is used to manage the shape of the point
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('scatter plot')
plt.show()
```

```python
# 3D plot
x =np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax=fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c ='red',alpha = .9 )
plt.xlabel('x axis')
plt.ylabel('Y axis')

plt.show()
```

```python
#piechart

langs = ['c' , 'python', 'java']
studnets = [20, 100, 40]

plt.pie(studnets, labels =langs , autopct = '%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
my_explod = (0.0 , 0.2, 0)

langs = ['c++' , 'python', 'java']
studnets = [20, 100, 40]

plt.pie(studnets, labels =langs , autopct = '%1.1f%%', explode = my_explod, shadow = True)
plt.title('Pie chart')
plt.show()
```
